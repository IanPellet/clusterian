```python
def FLAME_testparm(data, knn_, itn_):
    res_ = []
    for i in range(len(knn_)):
        clusterer = FLAME(cluster_neighbors=knn_[i], iteration_neighbors=itn_[i], verbose=0)
        clusterer = clusterer.fit(data)
        res_.append(clusterer.labels_)
        print("\r",f"{(i+1)/len(knn_)*100:.2f}","%",end="\r")

    nClust_ = [len(np.unique(r)) for r in res_]
    plt.plot(knn_, nClust_)
    return res_
```

High Kurtosis indicates a distribution with a long tail, producing much outliers. Here a high Kurtosis means that each cluster is well defined by a few genes and that most of genes are not included into the cluster. To define the genes included into the cluster, the False Discovery Rate (FDR) is computed.

```python
def getModules(fdr, FDRmax):
    if not isinstance(fdr,np.ndarray):
        fdr = np.array(fdr)
    modules = pd.DataFrame((fdr<FDRmax).astype(int), index=g2v.index)  
    return modules
```

```python
FDRmax_ = np.logspace(-15,-2,50)
rep = 100
    
FOM_ = []
aFOM_ = []

for i in range(len(FDRmax_)):
    fom = []
    afom = []
    for j in range(rep):
        fomj, afomj = utilsClust.FOM(pd.DataFrame(fdr), getModules, sk=False, FDRmax=FDRmax_[i])
        fom.append(fomj)
        afom.append(afomj)
        print("\r",f"{(i/len(FDRmax_)+(j+1)/(rep*len(FDRmax_)))*100:.2f}","%",end="\r")
    FOM_.append(statistics.median(fom))
    aFOM_.append(statistics.median(fom))
    print("\r",f"{(i+1)/len(FDRmax_)*100:.2f}","%",end="\r")
```

```python
plt.plot(FDRmax_, FOM_)
plt.xscale("log")
plt.xlabel("FDRmax")
plt.ylabel("FOM")
```
